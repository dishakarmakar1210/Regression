```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
data
```

```python
print(data.DESCR)
```

```python
#target is to predict the median house value for California districts, expressed in hundreds of thousands of dollars.
data.keys()
```

```python
data.target_names
```

```python
data.data
```

```python
data.feature_names
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
data.target_names
#data.target_names is an array of the target names. In this case, it is a single value: "MedHouseVal" (Median House Value).
```

```python
df['price'] = data.target
df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#scaling
#fit transform on training data and transform both training and test data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train[:5]
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
model = LinearRegression()
model
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse) 
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse, rmse, r2, mae
```

```python
plt.scatter(y_test, y_pred, color='blue', label='Actual')
```

```python
error = y_test - y_pred
sns.distplot(error, bins=30, kde=True, color='red')
```

```python
#pickle: helps store python object in a disk
#pickle is a python library that helps to serialize and deserialize python objects. It is used to save the model in a file and load it later for predictions.
#serialize: convert a python object into a byte stream
#deserialize: convert a byte stream back into a python object
import pickle
```

```python
model
```

```python
pickle.dump(model, open('model.pkl', 'wb')) #save the model in a file
#python's object (model) is converted in to bytes
```

```python
model = pickle.load(open('model.pkl', 'rb')) #load the model from the file
```
