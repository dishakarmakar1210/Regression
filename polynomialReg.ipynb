```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate same synthetic data or rreproducibility of results
np.random.seed(1)
```

```python
X = np.random.rand(100, 1) * 2
X
```

```python
#generate synthetic y
y = 4 + 3*X+ 1.5*X**2 + np.random.randn(100, 1)
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
poly
```

```python
X_poly_train = poly.fit_transform(X_train)
```

```python
X_poly_train 
```

```python
from sklearn.linear_model import LinearRegression
poly_reg = LinearRegression()
poly_reg
```

```python
poly_reg.fit(X_poly_train, y_train)
poly_reg
```

```python
poly_reg.coef_
```

```python
poly_reg.intercept_
```

```python
y_poly_reg = poly_reg.predict(X_poly_train)
y_poly_reg
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_train, y_poly_reg)
mse
```

```python
plt.scatter(X_train, y_train, color='blue', label='Training data')
```

```python
plt.scatter(X_train, y_train, color='blue', label='Training data')

X_range = np.linspace(0, 2, 100).reshape(-1, 1)

X_range_poly = poly.transform(X_range)
plt.plot(X_range, poly_reg.predict(X_range_poly), color='red', label='Polynomial regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression Fit')
plt.legend()
plt.show()
```
