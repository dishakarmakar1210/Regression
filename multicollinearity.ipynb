```python
#multicollinearity-- when two or more independent variables are highly correlated with each other
#when two features- correlation-- heatmap
#more than two features -- multicollinearity-- clustermap, VIF
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
```

```python
#clustermap-- shows group cluster relationship
plt.figure(figsize=(10, 8))
sns.clustermap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', figsize=(10, 8))
plt.title('Clustermap of Feature Correlation')
plt.show()
```

```python
#problem above -- we see that the relationship among features together but not quantitatively
#we need vif
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif["Feature"]= df.columns
```

```python
vif["VIF"] = [variance_inflation_factor(df.values, i) for i in range (len(df.columns))]
```

```python
vif
```

```python
# if vif is more than 10 then drop columns one by one
df1 = df.copy()
```

```python
#highest is longitude, we will drop it and then again calculate vif
df1.drop("Longitude", axis=1, inplace =True)
df1
```

```python
vif = pd.DataFrame()
vif["Feature"]= df1.columns
vif["VIF"] = [variance_inflation_factor(df1.values, i) for i in range (len(df1.columns))]
vif
```

```python
#drop avg rooms
df1.drop("AveRooms", axis = 1, inplace =True)
df1
```

```python
vif = pd.DataFrame()
vif["Feature"]= df1.columns
vif["VIF"] = [variance_inflation_factor(df1.values, i) for i in range (len(df1.columns))]
vif
```

```python
#drop latitude
df1.drop("Latitude", axis = 1, inplace =True)
df1
```

```python
vif = pd.DataFrame()
vif["Feature"]= df1.columns
vif["VIF"] = [variance_inflation_factor(df1.values, i) for i in range (len(df1.columns))]
vif
```

```python
#all vif are below 10 so we can build our model
df1
```

```python
X = df1.iloc[:, :-1] #IV
y = df1.iloc[:, -1] #target variable
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
#scaling -- option
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred= model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
#rmse 
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
score = r2_score(y_test, y_pred)
```
